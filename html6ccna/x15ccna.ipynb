````python
import re

def mFromPdf(l1, l2):
    '''
    '1','2.' => '1 2.'
    '''
    l1.append('11')
    l3 = []
    e0 = l1[0]
    for i in range(1, len(l1)):
        e = l1[i]
        match1 = re.search('[\w\,]$', e0)
        match2 = re.search('^[-#]', e)
        match3 = re.search('^[-#]', e0)
        match = match1 and not match2 and not match3
        if match:
            e0 += ' '+ e
        else:
            l3.append(e0)
            e0 = e
    for e in l3:
        l2.append(e)
    pass

def mToSentence(l1, l2):
    l3 = []
    
    
    for e in l1:
        match = re.search('^[#-]', e)
        if match:
            l3.append(e)
        else:
            l3.append('')
            m = re.split(r'(?<=[^A-Z].[.?!]) +(?=[A-Z])', e)
            l3 += m
            pass
    for e in l3:
        l2.append(e)
    pass

s = mReadFile('x15t3a.txt')
l = []
mList(s, l)
l2 = []
mFromPdf(l, l2)
l3 = []
mToSentence(l2, l3)
s2 = '\n'.join(l3)
# s2 = '\n'.join(l2)
mWriteFile('x15t3b.txt', s2)
# print(s[:320])
print(len(l2), len(l3))
````
