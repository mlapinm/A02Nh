## general

````python
from googletrans import Translator, constants
from pprint import pprint
# init the Google API translator
translator = Translator()

def translate(text):
    translation = translator.translate(text, src="en", dest="ru")
#     print(f"{translation.origin} ({translation.src}) --> {translation.text} ({translation.dest})")    return translation
    return translation.text

text = "This is a pen."    
translation = translate(text)   
print(translation)

def mReadFile(nameFile):
    s = ''
    with open(nameFile, encoding="utf8") as file:
        s = file.read()
    return s

def mWriteFile(nameFile, s):
    with open(nameFile, 'w', encoding="utf8") as file:
        file.write(s)
    return 

def translateListSmall(lien):
    sen = '\n'.join(lien)
    sru = translate(sen)
    liru = sru.split('\n')
    lienru = []
    for i in range(len(liru)):
        lienru.append(f'{lien[i]}|{liru[i]}')
    return lienru

def mList(s, l):
    '''
    s1\ns2 => [s1, s2]
    '''
    l2 = s.split('\n')
    l2 = list(map(lambda x: x.strip(), l2))
    for e in l2:
        l.append(e)
    pass


l = ['hello' , 'world']
l2 = translateListSmall(l)
print(l)
print(l2)
````

````python
import re

def mFromPdf(l1, l2):
    '''
    '1','2.' => '1 2.'
    '''
    l1.append('11')
    l3 = []
    e0 = l1[0]
    for i in range(1, len(l1)):
        e = l1[i]
        match1 = re.search('\w$', e0)
        match2 = re.search('^[-#]', e0)
        match = match1 and not match2
        if match:
            e0 += ' '+ e
        else:
            l3.append(e0)
            e0 = e
    for e in l3:
        l2.append(e)
    pass

def mToSentence(l1, l2):
    l3 = []
    
    
    for e in l1:
        match = re.search('^[#-]', e)
        if match:
            l3.append(e)
        else:
            l3.append('')
            m = re.split(r'(?<=[^A-Z].[.?!]) +(?=[A-Z])', e)
            l3 += m
            pass
    for e in l3:
        l2.append(e)
    pass

s = mReadFile('x15t3a.txt')
l = []
mList(s, l)
l2 = []
mFromPdf(l, l2)
l3 = []
mToSentence(l2, l3)
s2 = '\n'.join(l3)
mWriteFile('x15t3b.txt', s2)
# print(s[:320])
print(len(l3))
````
